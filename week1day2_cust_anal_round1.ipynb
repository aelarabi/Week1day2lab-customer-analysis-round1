```python
# FILE RENAMING
file1=file1.rename(columns={'customer lifetime value':'customer_lifetime_value','monthly premium auto':'monthly_premium_auto','number of open complaints':'number_of_open_complaints','policy type':'policy_type','vehicle class':'vehicle_class','total claim amount':'total_claim_amount'})
display(file1.columns)
file2=file2.rename(columns={'customer lifetime value':'customer_lifetime_value','monthly premium auto':'monthly_premium_auto','number of open complaints':'number_of_open_complaints','policy type':'policy_type','vehicle class':'vehicle_class','total claim amount':'total_claim_amount'})
display(file2.columns)
file3=file3.rename(columns={'customer lifetime value':'customer_lifetime_value','monthly premium auto':'monthly_premium_auto','number of open complaints':'number_of_open_complaints','policy type':'policy_type','vehicle class':'vehicle_class','total claim amount':'total_claim_amount'})
display(file3.columns)
```

```python
# FILE COLUMN REORDERING/REARRANGING:
file2=file2[['customer', 'st', 'gender', 'education', 'customer_lifetime_value','income', 'monthly_premium_auto', 'number_of_open_complaints','policy_type', 'vehicle_class', 'total_claim_amount']]
file3=file3[['customer', 'st', 'gender', 'education', 'customer_lifetime_value','income', 'monthly_premium_auto', 'number_of_open_complaints','policy_type', 'vehicle_class', 'total_claim_amount']]
```

```python
# TO CHECK COLUMN LABELS ARE IDENTICAL PRIOR TO CONCATENATE
file1.columns==file2.columns
```

```python
#CONCATENATE CODE
file123combine=pd.concat([file1,file2,file2], axis=0)
display(file123combine.shape)
display(file123combine)
```

```python
#RESET INDEX
file123combine=file123combine.reset_index(drop=True)
display(file123combine)
```

```python
#CHECK COLUMN TYPE USING INFO AND DTYPE 
display(file123combine.info())
display(file123combine.dtypes)
display(file123combine.columns)
```

```python
#Deleting columns
file123combineclean=file123combine.drop(['education','number_of_open_complaints'], axis=1)
```

```python
display(file123combineclean)
```

```python
#Converts the data from object to float and strips off uwanted symbol such as % that create errors
file123combineclean['customer_lifetime_value'] = file123combineclean['customer_lifetime_value'].str.rstrip('%').astype('float') / 100.0
```

```python
display(file123combineclean)
```

```python
#CHECK AND REMOVE DUPLICATES
file123combineclean_ndup=file123combineclean.copy()
file123combineclean_ndup=file123combineclean_ndup.drop_duplicates()
display(file123combineclean_ndup)
```

```python
#FILTER IN CLIENTS WITH INCOME > 0 AND OUT FILTER OUT CUSTOMERS WITH INCOME =< 0
file123labfinal=file123combineclean_ndup[file123combineclean_ndup['income']>0]
```
